```python
import urllib.request 
from bs4 import BeautifulSoup

```

```python
class scrapper:
    
    def __init__(self, site):
        self.site=site
        
    def newsscrape(self):
        r = urllib.request.urlopen(self.site)
        html=r.read()
        parser="html.parser"
        sp=BeautifulSoup(html,parser)
        for tag in sp.find_all("a"):
            url = tag.get("href")
            if url is None:
                continue
            if "html" in url:
                print("\n" + url)
                
    def textscrape(self):
        r = urllib.request.urlopen(self.site)
        html=r.read()
        parser="html.parser"
        sp=BeautifulSoup(html,parser)
        print(sp.get_text())
        
    def tablescrape(self):
        container=[]
        r = urllib.request.urlopen(self.site)
        html=r.read()
        parser="html.parser"
        sp=BeautifulSoup(html,parser)
        for i in sp.find_all("td"):
            container.append(i.get_text())
            #print(i.get_text() + "\n")
        return container
    
    def divscrape(self):
        container=[]
        r = urllib.request.urlopen(self.site)
        html=r.read()
        parser="html.parser"
        sp=BeautifulSoup(html,parser)
        for i in sp.find_all("div"):
            container.append(i.get_text())
        return container
            
            
        
    
          
```

```python
class organizer:
    
    def __init__(self, ddata, colnum):
        self.ddata=ddata
        self.colnum=colnum
   

    def columner(self):
        iterdata=iter(self.ddata)
        dlen=len(self.ddata)
        dlist=[]
        for i in range (0,self.colnum):
            dlist.append([])
        for x in range(0,dlen,self.colnum):
            for i in range(0,self.colnum):
                dlist[i].append(next(iterdata))
        return dlist

    
    def htmlgrapher(self):
        dlen=len(self.ddata)
        dlist=self.columner()
        htmlstr="<table> \n"
        rlen=int(dlen/self.colnum)
        for i in range(0,rlen):
            for x in range(0,self.colnum):
                if x==0:                
                    htmlstr+="<tr> \n" + "<td>" + str(dlist[x][i]) + "</td> \n"
                elif x==self.colnum:
                    htmlstr+= "<td>" + str(dlist[x][i]) + "</td>" + "\n </tr>"
                else:
                    htmlstr+="<td>"+ str(dlist[x][i]) + "</td>\n"
        htmlstr+="</table>"
        return print(htmlstr)
            
                    
                    
        
```
